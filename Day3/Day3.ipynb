```python
import time

# get input
with open('input.txt') as f:
    lines = f.readlines()
    
cleanedinput = [x.strip() for x in lines]
```

```python
cleanedinput = ['987654321111111','811111111111119','234234234234278','818181911112111']
cleanedinput
```

```python
# For the solution to part 1, set this to 2
digitswanted = 12
joltages = []

for line in cleanedinput:
    # Prepare empty string, allows easy adding of ints to string
    joltage = ''
    # Loop over all digits wanted, 
    for digit in range(1,digitswanted+1):
        # Get the max value in the list of digits, while ignoring the last X so we ensure 12 digit long number
        # the inline if is to account for list[:0], which doesnt work
        digitsignored = digitswanted-digit
        value = max(list(line[:-digitsignored])) if digitsignored != 0 else max(list(line))
        
        joltage += value
        
        # Truncate the string up until the found value
        for i in range(len(line)):
            if line[i] == value:
                line = line[i+1:]
                break
    
    joltages.append(int(joltage))

print(sum(joltages))
```
