```python
# For testing it's nice to have the example values, obviously dont run this if you want to use your own input
lines = "11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"
cleanedinput = []
for pair in lines.split(','):
    vals = pair.split('-')
    cleanedinput.append((vals[0], vals[1]))
cleanedinput
```

```python
invalids = []

for pair in cleanedinput:
    for ID in range(int(pair[0]), int(pair[1])+1):
        # Check for repeating numbers
        ID = str(ID)
        middle = len(ID)/2
        
        # Check if the string has an even length
        if int(middle) - middle == 0:
            # Check if first half equals last half
            if ID[:int(middle)] == ID[int(middle):]:
                invalids.append(int(ID))            
sum(invalids)
```

```python
start = time.time()
invalids = set()
# Pre-declaring a set and clearing it every time is quicker than declaring a new set every time
sections = set()

for pair in cleanedinput:
    for ID in range(int(pair[0]), int(pair[1])+1):
        # Check for repeating numbers
        ID = str(ID)
        length = len(ID)
        # Check all possible divisors
        for divisor in range(1, length):
            # Check if divides in equal parts
            if length % divisor == 0:
                # Use a set to see if all subparts of the ID are the same
                for i in range(int(length/divisor)):
                    sections.add(ID[i*divisor:(i+1)*divisor])
                if len(sections) == 1:
                    invalids.add(int(ID))
                sections.clear()
    
# Convert to set to remove duplicates
print(sum(invalids))

end = time.time()
print(f"Total runtime of the program is {end - start} seconds")
```
