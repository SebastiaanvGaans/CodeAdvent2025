```python
# get input
with open('day1_input.txt') as f:
    lines = f.readlines()
```

## Part 1

```python
timeszero = 0
position = 50
for line in lines:
#     clean and convert
    position += int(line.replace('\n','').replace('R','+').replace('L','-'))
    
#     loop around from 99 to 0 and otherwise
    position = position % 100
    
#     increment counter if at exactly 0
    if position==0:
        timeszero += 1
        
print('times at exactly 0 =', timeszero)
```
