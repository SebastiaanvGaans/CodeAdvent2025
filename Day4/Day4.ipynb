```python
import time
import copy
import pprint

# get input
with open('input.txt') as f:
    lines = f.readlines()

# strip and convert to list, list allows for replacing individual elements of a string
cleanedinput = [list(x.strip()) for x in lines]
```

```python
counter = 0
height = len(cleanedinput)
width = len(cleanedinput[0])

for cur_x in range(0, width):
    for cur_y in range(0, height):
        # If a . is here, continue on to next
        if cleanedinput[cur_y][cur_x] == '.':
            continue
            
        neighbours = 0
        # Loop over all possible neighbours
        for x in range(-1,2):
            for y in range(-1,2):
                # A tile is not its own neighbour
                if x == y == 0:
                    continue
                check_x = cur_x + x
                check_y = cur_y + y
                
                # Checks to see if its still in bounds
                if check_x < 0 or check_x >= width:
                    continue
                if check_y < 0 or check_y >= height:
                    continue
                
                # If a neighbour is found, increment counter
                if cleanedinput[cur_y+y][cur_x + x] != '.':
                    neighbours += 1
        if neighbours < 4:
            counter += 1
            # Mark the ones we can take with an X. Purely visual, does nothing for the function
            cleanedinput[cur_y][cur_x] = 'X'
            
print(counter, '\n')
for line in cleanedinput:
    print(''.join(line))
```

```python
def removepaper(matrix):
    counter = 0
    height = len(matrix)
    width = len(matrix[0])
    
    # Make a copy that we can return without altering the input 
    output = copy.deepcopy(matrix)
    
    for cur_x in range(0, height):
        for cur_y in range(0, width):
            # If a . is here, continue on to next
            if matrix[cur_y][cur_x] == '.':
                continue
            
            neighbours = 0
            # Loop over all possible neighbours
            for x in range(-1,2):
                for y in range(-1,2):
                    # A tile is not its own neighbour
                    if x == y == 0:
                        continue
                    check_x = cur_x + x
                    check_y = cur_y + y
                    
                    # Checks to see if its still in bounds
                    if check_x < 0 or check_x >= width:
                        continue
                    if check_y < 0 or check_y >= height:
                        continue
                    
                    # If a neighbour is found, increment counter
                    if matrix[cur_y+y][cur_x + x] != '.':
                        neighbours += 1
            if neighbours < 4:
                counter += 1
                # Mark the ones we have taken with a . , giving us the matrix for the next level,
                output[cur_y][cur_x] = '.'    
    return counter, output
```

```python
# Make a copy so I dont need to reload cleanedinput every time i run this block
matrix = copy.deepcopy(cleanedinput)

total = 0
counter = 1

while counter >0:
    counter, matrix = removepaper(matrix)
    total += counter
print(total)
```
