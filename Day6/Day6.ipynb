```python
from functools import reduce
import operator

# get input
with open('input.txt') as f:
    lines = f.readlines()

# strip and convert to list, list allows for replacing individual elements of a string
cleanedinput = [x.strip() for x in lines]
```

```python
cleanedinput = [
'123 328  51 64 ',
' 45 64  387 23 ',
'  6 98  215 314',
'*   +   *   +',
]
```

# Part 1

```python
# Turn the input into a matrix and filter out any empty strings due to double/triple spaces
matrix = []
for i in range(len(cleanedinput)):
    split = cleanedinput[i].split(' ')
    filtered = [x for x in split if x != '']
    matrix.append(filtered)


width = len(matrix[0])
height = len(matrix)-1
sumtotal = 0

for i in range(width):
    # Fetch what operator is used for this column and prepare the total value depending on the operator
    oper = matrix[-1][i]
    total = 0 if oper == '+' else 1
    
    # Sum up all values or get the sumproduct depending on the operator
    for j in range(height):
        if oper == '+':
            total += int(matrix[j][i])
        elif oper == '*':
            total *= int(matrix[j][i])
    sumtotal += total
print(f'The grand total is: {sumtotal}')
```

# Part 2

```python
# Clean and seperate all operators
operators = [x for x in cleanedinput[-1].split(' ') if x != '']
values = []
test = []

width = len(cleanedinput[0])
height = len(cleanedinput) -1

for i in range(width):
    # Loop over each column in the input and convert to numbers.
    column = ''
    for j in range(height):
        column += cleanedinput[j][i]
    column = column.replace(' ', '')
    
    # In the case of a seperator end the previous set of numbers and start a new one.
    if column == '':
        values.append(test)
        test = []
    else:
        test.append(int(column))
# Dont forget to include the last set which doesnt have a seperator.
values.append(test)

# Sum up or multiply each set of values and add to the total. python has no prod function so we use reduce.
sumtotal = 0
for i in range(len(values)):
    if operators[i] == '+':
        sumtotal += sum(values[i])
    elif operators[i] == '*':
        sumtotal += reduce(operator.mul, values[i], 1)
print(f'The grand total according to cephalopod math is: {sumtotal}')
```
