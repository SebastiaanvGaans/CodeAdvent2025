```python
# get input
with open('input.txt') as f:
    lines = f.readlines()

# strip and convert to list, list allows for replacing individual elements of a string
cleanedinput = [x.strip() for x in lines]
```

```python
cleanedinput = [
    '3-5',
    '10-14',
    '16-20',
    '12-18',
    '',
    '1',
    '5',
    '8',
    '11',
    '17',
    '32']
```

```python
raw_ranges = [x for x in cleanedinput if '-' in x]
ingredients = [int(x) for x in [x for x in cleanedinput if '-' not in x][1:]]

ranges = []
for IDrange in raw_ranges:
    start, stop = IDrange.split('-')
    ranges.append([int(start), int(stop)])
```

```python
fresh = 0

for ID in ingredients:
    # Check all ranges until one is found that includes the item ID. 
    # If one is found increment counter, break and continue on with next item ID
    for pair in ranges:
        if pair[0] <= int(ID) <= pair[1]:
            fresh +=1
            break
print(f"Fresh ingredients: {fresh}")
```

```python
# sorting the list by the starting value of all id ranges
sortedranges = sorted(ranges, key=lambda x: x[0])

# This removes any overlap between ranges
actualvalues = []
for val in sortedranges:
    if not actualvalues:
        actualvalues.append(val)
        continue
    # Check if there is any overlap with the previous range
    if val[0] <= actualvalues[-1][1]:
        # check if the current range is not a subset of the previous range
        if val[1] >= actualvalues[-1][1]:
            # If not a subset, extend previous range
            actualvalues[-1][1] = val[1]
    else:
        # If no overlap, add new range
        actualvalues.append(val)

total = 0
for x in actualvalues:
    total += x[1] - x[0] + 1
print(f"# Fresh ingredient IDs: {total}")
```
